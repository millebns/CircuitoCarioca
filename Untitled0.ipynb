```python
import matplotlib.pyplot as plt

# Dados para a legenda
numeros_rotulos = {local: str(i + 1) for i, local in enumerate(coordenadas.keys())}

# Criando uma nova figura para a legenda
plt.figure(figsize=(10, 6), facecolor='white')
plt.title("Legenda dos Pontos Turísticos", fontsize=16, color='black', loc='center', pad=20)

# Dividindo os itens em colunas
num_colunas = 3  # Número de colunas para exibir a legenda
itens_por_coluna = len(numeros_rotulos) // num_colunas + 1
colunas = [list(numeros_rotulos.items())[i:i + itens_por_coluna] for i in range(0, len(numeros_rotulos), itens_por_coluna)]

# Adicionando o texto da legenda para cada coluna
for col_index, coluna in enumerate(colunas):
    for linha_index, (local, numero) in enumerate(coluna):
        legenda_x = 0.1 + col_index * 0.3  # Ajuste da posição x para cada coluna
        legenda_y = 0.9 - linha_index * 0.05  # Ajuste da posição y para cada linha
        plt.text(legenda_x, legenda_y, f"{numero} - {local}", fontsize=10, color='black', ha='left')

# Removendo os eixos
plt.axis('off')

# Exibindo a figura
plt.show()

```
